```python
import os

def count_wav_files(input_path):
    total_wav_files = 0
    
    # Duyệt qua các thư mục DR1, DR2, ..., DR8
    for dr_folder in os.listdir(input_path):
        dr_path = os.path.join(input_path, dr_folder)
        if os.path.isdir(dr_path):
            # Duyệt qua các thư mục FAKS0, FCJF0, ...
            for speaker_folder in os.listdir(dr_path):
                speaker_path = os.path.join(dr_path, speaker_folder)
                if os.path.isdir(speaker_path):
                    # Đếm số lượng file .WAV.wav trong thư mục
                    wav_files = [file for file in os.listdir(speaker_path) if file.endswith(".WAV.wav")]
                    total_wav_files += len(wav_files)
    
    return total_wav_files

# Đường dẫn đầu vào
input_path = "../Data/TIMIT/data_v1/"

# Thực hiện đếm số lượng file .WAV.wav
total_files = count_wav_files(input_path)

# In ra kết quả
print(f"Tổng số file .WAV.wav trong 8 thư mục DR1 - DR8 là: {total_files}")
```

```python
import os
import shutil

def process_data(input_path, output_path):
    # Đếm số lượng file được gắn label
    count_labeled_files = 0
    
    # Biến để theo dõi index toàn cục
    global_index = 0
    
    # Duyệt qua các thư mục DR1, DR2, ..., DR8
    for dr_folder in os.listdir(input_path):
        dr_path = os.path.join(input_path, dr_folder)
        if os.path.isdir(dr_path):
            # Duyệt qua các thư mục FAKS0, FCJF0, ...
            for speaker_folder in os.listdir(dr_path):
                speaker_path = os.path.join(dr_path, speaker_folder)
                if os.path.isdir(speaker_path):
                    # Nhận diện giới tính từ chữ cái đầu tiên của tên thư mục
                    gender_label = "Female" if speaker_folder[0] == "F" else "Male"
                    
                    # Duyệt qua các tệp tin .wav trong thư mục
                    for wav_file in os.listdir(speaker_path):
                        if wav_file.endswith("WAV.wav"):
                            # Tăng giá trị của biến index toàn cục
                            global_index += 1
                            
                            # Tạo label mới cho tệp tin
                            new_label = f"{gender_label}_{wav_file.split('.')[0]}_{global_index}.wav"
                            
                            # Đường dẫn đầy đủ của tệp tin nguồn và đích
                            source_path = os.path.join(speaker_path, wav_file)
                            destination_path = os.path.join(output_path, new_label)
                            
                            # Copy và đổi tên tệp tin
                            shutil.copy(source_path, destination_path)
                            count_labeled_files += 1
                            print(f"Đã gắn label cho {os.path.abspath(source_path)} và sao chép vào {os.path.abspath(destination_path)}")

    return count_labeled_files
```

```python

# Đường dẫn đầu vào và đầu ra
input_path = "../Data/TIMIT/data_v1/"
output_path = "./dataset"

# Tạo thư mục đầu ra nếu chưa tồn tại
os.makedirs(output_path, exist_ok=True)

# Thực hiện xử lý dữ liệu và đếm số lượng file
count_files = process_data(input_path, output_path)

# In ra số lượng file được gắn label
print(f"Tổng số file được gắn label và sao chép là: {count_files}")
```

```python
import os

def count_files_with_labels(folder_path):
    # Kiểm tra xem thư mục tồn tại không
    if not os.path.exists(folder_path):
        print(f"Thư mục '{folder_path}' không tồn tại.")
        return

    female_count = 0
    male_count = 0

    # Duyệt qua tất cả các file trong thư mục
    for filename in os.listdir(folder_path):
        filepath = os.path.join(folder_path, filename)

        # Kiểm tra nếu là file và có đuôi là '.wav'
        if os.path.isfile(filepath) and filename.lower().endswith('.wav'):
            # Kiểm tra xem tên file có chứa 'Female' hay 'Male' không
            if 'female' in filename.lower():
                female_count += 1
            elif 'male' in filename.lower():
                male_count += 1

    # In kết quả
    print(f"Số lượng file Female: {female_count}")
    print(f"Số lượng file Male: {male_count}")

# Thực hiện đếm trong thư mục "./dataset"
count_files_with_labels("./dataset")
```
