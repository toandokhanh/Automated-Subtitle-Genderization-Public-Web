```python
import os
import shutil

def process_files(input_path, output_path):
    # Kiểm tra và tạo thư mục đầu ra nếu chưa tồn tại
    if not os.path.exists(output_path):
        os.makedirs(output_path)

    # Lặp qua tất cả các thư mục trong đường dẫn đầu vào
    for actor_folder in os.listdir(input_path):
        actor_path = os.path.join(input_path, actor_folder)

        # Kiểm tra xem có phải là thư mục không
        if os.path.isdir(actor_path):
            # Xác định giới tính dựa trên số trong tên thư mục
            gender = "Male" if int(actor_folder.split('_')[-1]) % 2 == 1 else "Female"

            # Lặp qua tất cả các tệp wav trong thư mục của diễn viên
            for wav_file in os.listdir(actor_path):
                if wav_file.endswith(".wav"):
                    # Tạo tên mới cho tệp wav
                    new_name = f"{gender}_{actor_folder}_{wav_file}"

                    # Đường dẫn đầy đủ của tệp cũ và mới
                    old_path = os.path.join(actor_path, wav_file)
                    new_path = os.path.join(output_path, new_name)

                    # Sao chép tệp vào thư mục đầu ra
                    shutil.copy(old_path, new_path)
                    print(f"Processed: {old_path} -> {new_path}")
```

```python
# Đường dẫn đầu vào và đầu ra của bạn
input_folder = "../../Data/RAVDESS"
output_folder = "dataset"

# Gọi hàm xử lý tệp
process_files(input_folder, output_folder)
```
